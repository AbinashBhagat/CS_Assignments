```python
# Q3. Given a number and ith bit, change the ith bit of that number to 1.
num = int(input("Number:"))
i = int(input('i:'))
mag_num = 1 << i
org_num = mag_num | num
bit = 0 if (num & mag_num > 0) else 1
print("""{0}th bit of number {1} changed to {2},
therefore the number {1} is changed to {3}""".format(i, num, bit, org_num))
print(org_num)
```
